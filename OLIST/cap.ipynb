```python
# Find the rows where "Aspect:" and "Sentiment:" are not present in the "judge" column
filtered_data = data[~(data['judge'].str.contains('Aspect:') & data['judge'].str.contains('Sentiment:'))]

data = data.drop(filtered_data.index)
data = data.reset_index(drop=True)
```

```python
temp  = data
```

```python
filtered_data = []

for index, row in temp.iterrows():
    aspects = row['judge'].count("Aspect:")
    sentiments = row['judge'].count("Sentiment:")
    if aspects == 1 and sentiments == 1:
        filtered_data.append(row)
filtered_df = pd.DataFrame(filtered_data)
```

```python
temp_filtered_df = filtered_df
```

```python
import re
aspect_regex = r"Aspect:\s*([^\n]+)"
sentiment_regex = r"Sentiment:\s*([^\n]+)"

temp_filtered_df['aspect'] = temp_filtered_df['judge'].str.extract(aspect_regex, flags=re.IGNORECASE)
temp_filtered_df['sentiment'] = temp_filtered_df['judge'].str.extract(sentiment_regex, flags=re.IGNORECASE)

temp_filtered_df['aspect'] = temp_filtered_df['aspect'].str.strip()
temp_filtered_df['sentiment'] = temp_filtered_df['sentiment'].str.strip()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

temp_filtered_df['aspect'] = temp_filtered_df['aspect'].str.lower()
temp_filtered_df['aspect'] = temp_filtered_df['aspect'].apply(lambda x: re.sub(r'[^\w\s]', '', x))
temp_filtered_df['aspect'] = temp_filtered_df['aspect'].str.lstrip()
temp_filtered_df['aspect'] = temp_filtered_df['aspect'].apply(lambda x: re.sub(r'^[-\s]+', '', x))
temp_filtered_df['aspect'] = temp_filtered_df['aspect'].apply(lambda x: re.sub(r'^[0-9\W_]+', '', x))
temp_filtered_df['aspect'] = temp_filtered_df['aspect'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))
```

```python
temp_filtered_df.sample(3)
```

```python
aspect_counts = temp_filtered_df['aspect'].value_counts()
aspect_counts
```

```python
label_mapping = {
    r"(?i).*delivery.*": "delivery",
    r"(?i).*product.*": "product",
    r"(?i).*service.*": "service"
}
def map_labels(label):
    for pattern, group in label_mapping.items():
        if re.match(pattern, label):
            return group
    return "others"
temp_filtered_df['main_aspect'] = temp_filtered_df['aspect'].apply(map_labels)
main_aspect_counts = temp_filtered_df['main_aspect'].value_counts()
```

```python
mapping = {
    'delivery': ['delivery', 'shipping', 'logistics', 'courier', 'arrival', 'time', 'timeliness', 'delay'],
    'product': ['product', 'quality', 'features', 'performance','price','functionality', 'price'],
    'service': ['service', 'customer support', 'assistance', 'helpdesk', 'experience', 'purchase', 'support']
}

def map_aspect_to_category(aspect):
    aspect = aspect.lower()
    for category, keywords in mapping.items():
        for keyword in keywords:
            if keyword in aspect:
                return category
    return 'others'
df_2['main_aspect'] = df_2['aspect'].apply(map_aspect_to_category)

main_aspect_counts_others = df_2['main_aspect'].value_counts()
```
